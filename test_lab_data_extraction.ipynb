```python
import numpy as np
import pandas as pd
```

```python
lab_data = pd.read_csv('data/lab_data.csv')
```

```python
lab_data = lab_data.loc[lab_data['labresultoffset'] > -1]
lab_data = lab_data.loc[lab_data['labresultoffset'] <= 24*60]
display(lab_data)
```

```python
lab_data = lab_data[['patientunitstayid', 'labresultoffset', 'labname', 'labresult']].drop_duplicates()
lab_data.rename({'labresultoffset':'offset'})
display(lab_data)
```

```python
lab_data_piv = pd.pivot_table(lab_data, index=['patientunitstayid', 'labresultoffset'], columns='labname', values = 'labresult').reset_index()
display(lab_data_piv)
lab_data_piv.shape
```

```python
# TODO exclude missing lab features after selecting the correct patient set
lab_counts = lab_data_piv.groupby('patientunitstayid').count().reset_index().drop(columns=['labresultoffset'])
lab_counts = lab_counts[lab_counts.columns[lab_counts.sum() > lab_counts.shape[0] * 4 / 5]]
display(lab_counts)

lab_avgs = lab_data_piv.groupby('patientunitstayid').mean().reset_index().drop(columns=['labresultoffset'])
lab_avgs = lab_avgs[lab_counts.columns[lab_counts.sum() > lab_counts.shape[0] * 4 / 5]]
lab_avgs = lab_avgs.apply(lambda x: x.fillna(x.median()))
```
