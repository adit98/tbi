```python
X_ts = pd.read_csv('../data/loaded/processed/ts_processed.csv')
X_hr = X_ts[['patientunitstayid', 'offset_bin', 'hr']].pivot(index='patientunitstayid',
            columns='offset_bin', values='hr').reset_index()
X_resp = X_ts[['patientunitstayid', 'offset_bin', 'resp']].pivot(index='patientunitstayid',
            columns='offset_bin', values='resp').reset_index()
X_sao2 = X_ts[['patientunitstayid', 'offset_bin', 'sao2']].pivot(index='patientunitstayid',
            columns='offset_bin', values='sao2').reset_index()
X_gcs = X_ts[['patientunitstayid', 'offset_bin', 'gcs']].pivot(index='patientunitstayid',
            columns='offset_bin', values='gcs').reset_index()
X_gcs.iloc[:, 1:][X_gcs.iloc[:, 1:] > 6] = 6
X_gcs.iloc[:, 1:][X_gcs.iloc[:, 1:] < 1] = 1

display(X_gcs)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 5)
hrp = pca.fit_transform(X_hr.iloc[:, 1:])
respp = pca.fit_transform(X_resp.iloc[:, 1:])

sao2p = pca.fit_transform(X_sao2.iloc[:, 1:])
plt.figure(figsize = (8, 8))
plt.title("sao2 PCA 0")
plt.stem(pca.components_[0, :])

gcsp = pca.fit_transform(X_gcs.iloc[:, 1:])
plt.figure(figsize = (8, 8))
plt.title("GCS PCA 0")
plt.stem(pca.components_[0, :])


hr = pd.DataFrame(hrp)
hr.columns = ['hr0', 'hr1', 'hr2', 'hr3', 'hr4']
hr['patientunitstayid'] = X_hr['patientunitstayid']

resp = pd.DataFrame(respp)
resp.columns = ['resp0', 'resp1', 'resp2', 'resp3', 'resp4']
resp['patientunitstayid'] = X_resp['patientunitstayid']

sao2 = pd.DataFrame(sao2p)
sao2.columns = ['sao20', 'sao21', 'sao22', 'sao23', 'sao24']
sao2['patientunitstayid'] = X_sao2['patientunitstayid']

gcs = pd.DataFrame(gcsp)
gcs.columns = ['gcs0', 'gcs1', 'gcs2', 'gcs3', 'gcs4']
gcs['patientunitstayid'] = X_gcs['patientunitstayid']
display(resp)
```

```python
X = gcs.set_index('patientunitstayid').merge(hr.set_index('patientunitstayid'), how = 'left', left_index = True, right_index = True).merge(resp.set_index('patientunitstayid'), how = 'left',left_index = True, right_index = True).merge(sao2.set_index('patientunitstayid'), how = 'left',left_index = True, right_index = True)
Xn = X.merge(nc, how = 'left',left_index = True, right_index = True).merge(med,  how = 'left',left_index = True, right_index = True).merge(lab, how = 'left',left_index = True, right_index = True).merge(inf,  how = 'left',left_index = True, right_index = True).merge(dem,  how = 'left',left_index = True, right_index = True).merge(ap,  how = 'left',left_index = True, right_index = True)
Xn = Xn.reset_index()
```

```python
#out = GCS, los, death
out = "los"
Y = Xn.merge(mort)[out].values
Xns = Xn.merge(mort).iloc[:, :-4]
Xns = Xns.fillna(Xns.mean())
display(Y)
display(Xns)

X = Xns.drop('patientunitstayid', axis = 1).values
Xn = Xns.drop('patientunitstayid', axis = 1)
y = Y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import ElasticNet
import statsmodels.api as sm

if out == "los":
    print("LOS")
    y = y > 5
yclass = y
sss = StratifiedShuffleSplit(n_splits=20, test_size=0.3, random_state=6)
trainpredsss = []
testpredsss = []
trainlabsss = []
testlabsss = []
i = 0
scaler = StandardScaler()
coefs = np.zeros((20, X.shape[1]))
for train_index, test_index in sss.split(X, yclass):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    lg = LogisticRegression(max_iter=1000, penalty='elasticnet', solver='saga', C=4, class_weight = 'balanced')
    parameters = {'C':[.02, .05, .1], 'l1_ratio': [.9, .95, .97]}  
    clf = GridSearchCV(lg, parameters, cv = 3)
    clf.fit(X_train, y_train)
    coefs[i, :] = clf.best_estimator_.coef_
    trainpreds = clf.predict_proba(X_train)
    testpreds = clf.predict_proba(X_test)
    trainpredsss.append(trainpreds[:, 1])
    testpredsss.append(testpreds[:, 1])
    trainlabsss.append(y_train)
    testlabsss.append(y_test)
    i = i + 1
     
```

```python
meanCoefs = np.abs(np.median(coefs, axis = 0))
print(meanCoefs.shape)
coefInds = np.argsort(np.abs(meanCoefs))[::-1][:10]

plt.figure(figsize = (10, 10))
if out == "los":
    plt.ylabel("<-- short stay -- long stay -->", size = 16)
else:
    plt.ylabel("<-- good outcome -- bad outcome -->", size = 16)
plt.title("LOS coefficients")
plt.xticks(rotation = 20)
plt.boxplot(coefs[:, coefInds], labels = Xn.columns[coefInds])
plt.savefig("LOScoef.png", bbox_inches = 'tight')
```

```python
def reindex_thresh(tpr, fpr, thresh):
    # get thresh to reindex by
    thresh_lens = np.array([i.shape[0] for i in thresh])
    rein_thresh = thresh[np.argmax(thresh_lens)]

    # reindex all these values so we are using the same thresholds
    for i in range(len(thresh)):
        df = pd.DataFrame(columns=['fpr', 'tpr', 'thresh'])
        df['fpr'] = fpr[i]
        df['tpr'] = tpr[i]
        df['thresh'] = thresh[i]

        df = df.set_index('thresh').reindex(index=rein_thresh, method='nearest').reset_index()
        fpr[i] = df['fpr'][:, None]
        tpr[i] = df['tpr'][:, None]
        thresh[i] = df['thresh'][:, None]

    fpr = np.hstack(fpr).T
    tpr = np.hstack(tpr).T
    thresh = np.hstack(thresh).T
    
    return tpr, fpr, thresh
```

```python
from sklearn import metrics
plt.figure(figsize = (8, 8))
fprs = []
tprs = []
aucs = []
threshs = []
for i in np.arange(20):
    fpr, tpr, thresholds = metrics.roc_curve(trainlabsss[i], trainpredsss[i])
    auc = metrics.auc(fpr, tpr)
    print(fpr.shape)
    fprs.append(fpr)
    tprs.append(tpr)
    threshs.append(thresholds)
    aucs.append(auc)

aucs = np.array(aucs)
tpr, fpr, thresh = reindex_thresh(tprs, fprs, threshs)
avg_tpr = np.average(tpr, axis=0)
avg_fpr = np.average(fpr, axis=0)
avg_auc = np.average(aucs)
plt.plot(avg_fpr, avg_tpr, c = 'b', linewidth = 2)
error = np.std(tpr, axis = 0).astype(float)
plt.fill_between(avg_fpr, avg_tpr-error, avg_tpr+error, alpha = .2, color = 'b')
tauc = np.round(avg_auc, 4)

fprs = []
tprs = []
aucs = []
threshs = []
for i in np.arange(20):
    fpr, tpr, thresholds = metrics.roc_curve(testlabsss[i], testpredsss[i])
    auc = metrics.auc(fpr, tpr)
    print(fpr.shape)
    fprs.append(fpr)
    tprs.append(tpr)
    threshs.append(thresholds)
    aucs.append(auc)

aucs = np.array(aucs)
tpr, fpr, thresh = reindex_thresh(tprs, fprs, threshs)
avg_tpr = np.average(tpr, axis=0)
avg_fpr = np.average(fpr, axis=0)
avg_auc = np.average(aucs)
avg_auc = np.round(avg_auc, 4)
plt.plot(avg_fpr, avg_tpr, c = 'r', linewidth = 2)
error = np.std(tpr, axis = 0).astype(float)
plt.fill_between(avg_fpr, avg_tpr-error, avg_tpr+error, alpha = .2, color = 'r')

plt.xlabel("FPR", size = 16)
plt.ylabel("TPR", size = 16)
plt.ylim((0, 1))
plt.title("LOS ROC: Train Auc = " + str(tauc) + ", Test = " + str(avg_auc), size = 16)
plt.savefig("LOS ROC", bbox_inches = 'tight')
```
