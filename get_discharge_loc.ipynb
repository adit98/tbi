```python
import pandas as pd
import numpy as np
```

```python
dem = pd.read_csv('data/patient_demographics_data.csv')
```

```python
dem_mod = dem[['patientunitstayid', 'unitadmitsource', 'unitdischargeoffset', 'unitdischargelocation']]
display(dem_mod)
```

```python
dem_mod.unitdischargelocation.unique()
```

```python
replace_dict = {'Home': 6, 'Skilled Nursing Facility': 5, 'Nursing Home': 5, 'Rehabilitation': 5,
               'Other External': 5, 'Floor': 4, 'Other Internal': 4, 'Step-Down Unit (SDU)': 3,
               'Other Hospital': 3, 'ICU': 2, 'Other ICU': 2, 'Other ICU (CABG)': 2, 'Acute Care/Floor': 2,
               'Telemetry': 1, 'Operating Room': 1, 'Death': 0, 'Other': np.nan}

dem_mod_replaced = dem_mod.copy()
dem_mod_replaced['unitdischargelocation'].replace(replace_dict, inplace=True)
```

```python
display(dem_mod_replaced)
```

```python
dem_mod_replaced['unitdischargelocation'].fillna(lambda x: x.mean(), inplace=True)
```
